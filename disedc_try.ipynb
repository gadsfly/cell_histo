```python
from skimage import io, transform, exposure
import numpy as np

dapi_rgb = io.imread('PVN, MD, RE image61_R 91_Merged_ch00_SV.tif')
dapi = dapi_rgb[..., 2] if len(dapi_rgb.shape) == 3 else dapi_rgb

# Resize to ~1000px (DeepSlice uses 299x299 anyway)
scale = 1000 / max(dapi.shape)
dapi_small = transform.resize(dapi, (int(dapi.shape[0]*scale), int(dapi.shape[1]*scale)), 
                               preserve_range=True, anti_aliasing=True)

# Enhance & save as 8-bit PNG
p2, p98 = np.percentile(dapi_small, (2, 98))
dapi_8bit = exposure.rescale_intensity(dapi_small, in_range=(p2, p98), out_range=(0,255)).astype(np.uint8)
io.imsave('dapi_for_deepslice.png', dapi_8bit)
print("Done! Upload dapi_for_deepslice.png instead")
```
